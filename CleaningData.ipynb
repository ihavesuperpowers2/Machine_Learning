```python
filename = 'pg5200.txt'
file = open(filename, 'rt')
text = file.read()
file.close()
```

# This is data preprocessing manually.

```python
import re
words = re.split(r'\W+', text)
print(words[:100])
```

```python
print(text)
```

```python
!pip install nltk

import nltk
nltk.download('punkt')

#splitting to words

from nltk.tokenize import word_tokenize

tokenize = word_tokenize(text)
print(tokenize[:100])
```

# Using sent tokenize
To split by sentences.

```python
from nltk import sent_tokenize
token = sent_tokenize(text)
```

```python
print(token[:100])
```

# Filtering out puntuations


```python
words = [word for word in tokenize if word.isalpha()]
print(words[:100])
```

# Filtering out stopwords like 'the', 'and' etc.

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
print(stop_words)
```

```python
stop_words = set(stopwords.words('english'))
```

```python
stopwords = [w for w in words if w not in stop_words]
print(stopwords[:100])
```

# Stemming words i.e. reducing them to the roots.

```python
from nltk.stem.porter import PorterStemmer
```

```python
porter = PorterStemmer()
stemmed_words = [porter.stem(word) for word in tokenize]
```

```python
print(stemmed_words)
```
